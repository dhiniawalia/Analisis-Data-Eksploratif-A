### Deskripsi Kasus

Data yang digunakan merupakan data tingkat kemiskinan di Indonesia. Data ini mencakup beberapa indikator penting terkait kesejahteraan masyarakat, pembangunan manusia, kemiskinan, dan akses terhadap layanan dasar. Berikut adalah deskripsi dari setiap kolom yang ada di data tersebut:

1. Provinsi: Nama provinsi di mana kabupaten/kota tersebut berada. Dalam data ini, ada dua provinsi, yaitu Aceh dan Sumatera Utara.

2. Kab/Kota: Nama kabupaten atau kota yang merupakan unit administratif dalam provinsi.

3. Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen): Persentase penduduk yang hidup di bawah garis kemiskinan di masing-masing kabupaten/kota.

4. Rata-rata Lama Sekolah Penduduk 15+ (Tahun): Jumlah tahun rata-rata sekolah yang telah ditempuh oleh penduduk yang berusia 15 tahun ke atas di masing-masing wilayah.

5. Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun): Rata-rata pengeluaran per kapita (disesuaikan) di masing-masing kabupaten/kota dalam satuan ribu rupiah per orang per tahun.

6. Indeks Pembangunan Manusia (IPM): Indeks yang mengukur pencapaian rata-rata dalam tiga dimensi dasar pembangunan manusia, yaitu kesehatan (diukur dari umur harapan hidup), pendidikan (rata-rata lama sekolah dan harapan lama sekolah), dan standar hidup layak (diukur dari PDB per kapita).

7. Umur Harapan Hidup (Tahun): Rata-rata tahun seseorang diperkirakan akan hidup sejak lahir di setiap kabupaten/kota.

8. Persentase Rumah Tangga yang Memiliki Akses terhadap Sanitasi Layak: Persentase rumah tangga yang memiliki akses terhadap sanitasi layak di masing-masing kabupaten/kota.

9. Persentase Rumah Tangga yang Memiliki Akses terhadap Air Minum Layak: Persentase rumah tangga yang memiliki akses terhadap air minum layak di masing-masing kabupaten/kota.

10. Tingkat Pengangguran Terbuka: Persentase angkatan kerja yang menganggur di setiap kabupaten/kota.

11. Tingkat Partisipasi Angkatan Kerja: Persentase jumlah penduduk yang aktif dalam angkatan kerja di setiap kabupaten/kota.

12. PDRB atas Dasar Harga Konstan Menurut Pengeluaran (Rupiah): Produk Domestik Regional Bruto (PDRB) yang diukur atas dasar harga konstan (dengan tahun acuan tertentu) menurut pengeluaran, yang merepresentasikan output ekonomi di kabupaten/kota tersebut.

13. Klasifikasi Kemiskinan: Indikator biner yang menunjukkan apakah suatu wilayah diklasifikasikan sebagai "Miskin" (1) atau "Tidak Miskin" (0) berdasarkan berbagai indikator kemiskinan.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("C://Users//ASUS//Documents//Klasifikasi Tingkat Kemiskinan di Indonesia1.csv")
data.head()
```

```python
data['Klasifikasi Kemiskinan'].value_counts()
```

```python
# Menampilkan Dimensi Dataset
print('Dimensi data:', data.shape)
print("")

# Menampilkan Jumlah Dataset
print('jumlah dataset:'+ str(len(data)))
```

```python
# Menampikan Missing Value
data.isnull().sum()
```

```python
# Menghilangkan Missing Value
data.dropna(inplace=True)

# Menampilkan data yang telah dibersihkan dari missing value
data
```

```python
# Analisis Deskriptif
data.describe()
```

```python
# Menampilkan informasi variabel terkait tipe data
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Mendefinisikan kolom yang tidak akan diubah
columns_to_exclude = ['Provinsi', 'Kab/Kota']

# Mengubah tipe data ke integer untuk kolom selain yang dikecualikan
label_encoder = LabelEncoder()
for column in data.columns:
    if column not in columns_to_exclude:
        data[column] = label_encoder.fit_transform(data[column].astype(str))
```

```python
# Menampilkan informasi variabel terkait tipe data
data.info()
```

```python
# Histogram
data.hist(figsize=(15, 8))
plt.title('Pesebaran Data')
plt.tight_layout()
plt.show()
```

```python
# Mengganti 0 : Tidak Miskin dan 1 : Miskin
data['Klasifikasi Kemiskinan'] = data['Klasifikasi Kemiskinan'].map({0: 'Tidak Miskin', 1: 'Miskin'})

print('Grafik Distribusi Kemiskinan di Indonesia:')
print(data['Klasifikasi Kemiskinan'].value_counts())

plt.figure(figsize=(10, 6))
data['Klasifikasi Kemiskinan'].value_counts().plot(kind='bar')
plt.title('Klasifikasi Kemiskinan')
plt.xlabel('Jenis')
plt.ylabel('Jumlah')
plt.show()
```

```python
y_axis = data['Klasifikasi Kemiskinan'].value_counts()
my_labels = ['Tidak Miskin', 'Miskin']
color = ['#B5C0D0', '#EED3D9']
explode = [0, 0.2]

plt.pie(y_axis, labels=my_labels, explode=explode,
        colors=color, autopct='%1.1f%%')
plt.show()

print('\nDistribution of IPM Classification:')
print(data['Klasifikasi Kemiskinan'].value_counts())
```

```python
kota_miskin = data.loc[data["Klasifikasi Kemiskinan"] == "Miskin"].sort_values(by='Indeks Pembangunan Manusia', ascending=True)
y_a = kota_miskin["Indeks Pembangunan Manusia"]
x_a = kota_miskin["Kab/Kota"]
bar_colors = '#55AD9B'

plt.figure(figsize=(35, 6))
plt.bar(x_a, y_a, color=bar_colors)
plt.title('Kabupaten/Kota Yang Tergolong Miskin')
plt.xlabel('Kabupaten/Kota')
plt.ylabel('IPM')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(y=data["Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)"], hue=data["Klasifikasi Kemiskinan"])
plt.show()
sns.countplot(y=data["Indeks Pembangunan Manusia"], hue=data["Klasifikasi Kemiskinan"])
plt.show()
sns.countplot(y=data["Tingkat Pengangguran Terbuka"], hue=data["Klasifikasi Kemiskinan"])
plt.show()
```

```python
sns.distplot(data["Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)"])
plt.show()
sns.distplot(data["Indeks Pembangunan Manusia"])
plt.show()
sns.distplot(data["Tingkat Pengangguran Terbuka"])
plt.show()
```

```python
data1 = ['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)','Rata-rata Lama Sekolah Penduduk 15+ (Tahun)',
         'Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)','Indeks Pembangunan Manusia','Umur Harapan Hidup (Tahun)',
         'Persentase rumah tangga yang memiliki akses terhadap sanitasi layak','Persentase rumah tangga yang memiliki akses terhadap air minum layak',
         'Tingkat Pengangguran Terbuka','Tingkat Partisipasi Angkatan Kerja', 'PDRB atas Dasar Harga Konstan menurut Pengeluaran (Rupiah)']

# Memilih kolom yang ada di dalam list data1 dari DataFrame df
data_selected = data[data1]

# Menghitung korelasi dari kolom numerik yang dipilih
corr_matrix = data_selected.corr()

# Membuat heatmap dari matriks korelasi
sns.heatmap(corr_matrix, linewidths=.5, annot=True, cmap='RdYlGn')

# Mengatur ukuran plot
fig = plt.gcf()
fig.set_size_inches(15, 8)
plt.show()
```
